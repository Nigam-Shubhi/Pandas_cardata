```python
a["EngineSize"].fillna(a["EngineSize"].mean(),inplace=True)
a["Cylinders"].fillna(a["Cylinders"].mean(),inplace=True)
a["Horsepower"].fillna(a["Horsepower"].mean(),inplace=True)
a["MPG_City"].fillna(a["MPG_City"].mean(),inplace=True)
a["MPG_Highway"].fillna(a["MPG_Highway"].mean(),inplace=True)
a["Weight"].fillna(a["Weight"].mean(),inplace=True)
a["Wheelbase"].fillna(a["Wheelbase"].mean(),inplace=True)
a["Length"].fillna(a["Length"].mean(),inplace=True)
```

```python
a.drop_duplicates(inplace=True)
a
```
